```python
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset('car_crashes')
df.head()
```

```python
plt.scatter(df.speeding, df.alcohol)
plt.show()
```

```python
plt.scatter(df.speeding, df.alcohol)
#sns.set_style("whitegrid")
#sns.set_style("darkgrid")
#sns.set_style("white")
#sns.set_style("dark")
sns.set_style("ticks")
sns.despine()
plt.show()
```

```python
plt.scatter('speeding', 'alcohol', data=df)
sns.set_style("darkgrid", {'axes.edgecolor':'black'})
plt.show()
```

```python
sns.relplot(data=df, x="speeding", y="alcohol")
```

```python
conditions=[(df['total']<=10), (df['total']>10)&(df['total']<=15), (df['total']>15)&(df['total']<=20), (df['total']>20)]
values=['Low', 'Moderate', 'High', 'Very high']
df['type']=np.select(conditions, values)
df.head()
```

```python
conditions=[(df['ins_premium']<=1000), (df['ins_premium']>1000)]
values=['Low (B)', 'High (A)']
df['premium_type']=np.select(conditions, values)
df.head()
```

```python
sns.relplot(data=df, x="speeding", y="alcohol", hue="type")
```

```python
sns.relplot(data=df, x="speeding", y="alcohol", hue="premium_type", col="type", col_wrap=2)
```

```python
df.speeding.plot(kind="hist")
```

```python
sns.distplot(df['speeding'], kde=True, bins=25)
```

```python
sns.barplot(x='premium_type', y='alcohol', hue='type', data=df, palette='PuRd', order=['High (A)', 'Low (B)'], capsize=0.05, saturation=8, errcolor='gray', errwidth=2, ci='sd')
plt.legend()
plt.show()
```

```python
sns.violinplot(x='premium_type', y='alcohol', hue='type', data=df, palette='GnBu')
```

```python
sns.pointplot(x="type", y="alcohol", data=df)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=df)
sns.jointplot(x='speeding', y='alcohol', data=df, kind='hex')
sns.jointplot(x='speeding', y='alcohol', data=df, kind='kde')
```

```python
ax=sns.regplot(x='speeding', y='alcohol', data=df)
```

```python
corr=df.corr()
print(corr)
sns.heatmap(corr, annot=True, linewidth=1, cmap='YlGnBu')
```

```python
sns.set_style('ticks')
sns.pairplot(df, hue='type', diag_kind='kde', kind='scatter', palette='husl')
plt.show()
```
