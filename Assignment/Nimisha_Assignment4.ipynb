# Report best regression model


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
x=np.arange(0, 30).reshape((-1, 1))
y=np.array([3,4,5,7,10,8,9,10,10,23,27,44,50,63,67,60,62,70,75,88,81,87,95,100,108,135,151,160,169,179])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

LINEAR REGRESSION MODEL

```python
lr=LinearRegression()
lr.fit(x_train, y_train)
lr.intercept_, lr.coef_
```

```python
y_pred_lr=lr.predict(x_test)
y_pred_lr
```

STOCHASTIC GRADIENT DESCENT REGRESSOR

```python
sgdr=SGDRegressor(max_iter=1000, penalty=None, eta0=0.05)
sgdr.fit(x, y.ravel())
sgdr.intercept_, sgdr.coef_
```

```python
y_pred_sgdr=sgdr.predict(x_test)
y_pred_sgdr
```

POLYNOMIAL REGRESSION

```python
print(pd.DataFrame(x).head())
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
pr=PolynomialFeatures(degree=3)
x_train2=pr.fit_transform(x_train)
x_test2=pr.fit_transform(x_test)
```

```python
pr_clf=LinearRegression()
pr_clf.fit(x_train2, y_train)
y_pred_pr=pr_clf.predict(x_test2)
pr_clf.score(x_test2, y_test)
```

RIDGE REGRESSION

```python
rr=Ridge(alpha=1)
rr.fit(x_train, y_train)
y_pred_rr=rr.predict(x_test)
rr_poly=Ridge(alpha=1)
rr_poly.fit(x_train2, y_train)
y_pred_rr-rr_poly.predict(x_test2)
```

```python
print(format(rr.score(x_test, y_test)))
print(np.sum(rr.coef_!=0))
print(format(rr_poly.score(x_test2, y_test)))
print(np.sum(rr_poly.coef_ !=0))
```

LASSO REGRESSION

```python
lasso=Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
y_pred_lasso=lasso.predict(x_test)
```

```python
print(lasso.score(x_train, y_train))
print(lasso.score(x_test, y_test))
print(np.sum(lasso.coef_!=0))
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred_lr)), 
      np.sqrt(mean_squared_error(y_test, y_pred_sgdr)), 
      np.sqrt(mean_squared_error(y_test, y_pred_pr)), 
      np.sqrt(mean_squared_error(y_test, y_pred_rr)), 
      np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
```

```python
print(r2_score(y_test, y_pred_lr), 
      r2_score(y_test, y_pred_sgdr), 
      r2_score(y_test, y_pred_pr), 
      r2_score(y_test, y_pred_rr), 
      r2_score(y_test, y_pred_lasso))
```

Hence it can be concluded that the best regression model for the given dataset is the Polynomial Regression Model.
