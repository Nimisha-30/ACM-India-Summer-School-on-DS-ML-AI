```python
mystring=input("Enter string: ")
searchstring=input("Enter word to be searched: ")
searchword1=" "+searchstring+" "
searchword2=searchstring+" "
searchword3=" "+searchstring+"."
searchword4=" "+searchstring+","
count1=mystring.count(searchword1)
count2=mystring.count(searchword2)
count3=mystring.count(searchword3)
count4=mystring.count(searchword4)
#print(count1, count2, count3, count4)
count=count1+count2+count3+count4
print("Number of occurences of '", searchstring, "' in the given string is ", count)
```

```python
from statistics import mean
fibseries=[0, 1]
def fibonacci_series_for():
  for i in range(597):
    nextfibelement=fibseries[i]+fibseries[i+1]
    fibseries.append(nextfibelement)
  print("Fibonacci series: ", fibseries)
  fibratioseries=[]
  for i in range(1, len(fibseries)-1):
    ratio=fibseries[i+1]/fibseries[i]
    fibratioseries.append(ratio)
  print("Ratio of consecutive elements of fibonacci series: ", fibratioseries)
  goldenratio=mean(fibratioseries)
  print("Golden ratio: ", round(goldenratio, 3))
def fibonacci_series_while():
  j=0
  while j<598:
    nextfibelement=fibseries[j]+fibseries[j+1]
    fibseries.append(nextfibelement)
  print("Fibonacci series: ", fibseries)
  fibratioseries=[]
  j=1
  while j<len(fibseries):
    ratio=fibseries[j+1]/fibseries[j]
    fibratioseries.append(ratio)
  print("Ratio of consecutive elements of fibonacci series: ", fibratioseries)
  goldenratio=mean(fibratioseries)
  print("Golden ratio: ", round(goldenratio, 3))
fibonacci_series_for()
fibonacci_series_while()
```
