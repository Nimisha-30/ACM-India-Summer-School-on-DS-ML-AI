```python
import seaborn as sns
data=sns.load_dataset('tips')
data.head()
```

Shown below is a scatter plot of day against tip.

```python
sns.scatterplot(x='tip', y='day', data=data)
```

Shown below is a histogram plot of total bill of the customers.

```python
sns.displot(data['total_bill'], kde=True)
```

Shown below is a line plot of the relationship between total bill and gender.

```python
sns.lineplot(x='sex', y='total_bill', data=data)
```

Shown below are 4 plots separated by days and sex of total bill and tip.

```python
sns.relplot(x='total_bill', y='tip', kind='scatter', hue='sex', col='day', col_wrap=4, data=data)
```

Shown below is a point plot of tip against day.

```python
sns.pointplot(x='day', y='tip', data=data)
```
